# Passo 1: Coleta e preparação dos dados

```python
import pandas as pd

# Importar dataset
tabela = pd.read_csv("ecommerce_sales_data.csv")

# Explorar a estrutura do dataset
display(tabela)
```

Tratar valores ausentes, inconsistentes e seus tipos

```python
display(tabela.info())
```

Para realizar a análise de sazonalidade, foi incluído para análise uma coluna do mês da venda e o dia da semana

```python
# Converter a coluna de data para formato datetime - Assim dá para trabalhar com as informações dela
tabela['date'] = pd.to_datetime(tabela['date'])

# Adicionando coluna de mês
tabela["month"] = tabela["date"].dt.month

# Adicionando coluna de dia da semana
tabela["day_of_week"] = tabela["date"].dt.day_name()
```

```python
display(tabela)
```

# Passo 2: Análise exploratória dos dados (EDA - Exploratory Data Analusis)

Analisar tendências gerais

```python
# Contagem de vendas por categoria
tabela["category"].value_counts()
```

```python
# Contagem de vendas por região
tabela["region"].value_counts()
```

```python
# Agrupar vendas por mês
tabela_vendas_mes = tabela.groupby("month")["sales"].sum().reset_index()
```

```python
import plotly.express as px

# Criando um gráfico de pizza para as vendas do mês
grafico_pizza = px.pie(tabela_vendas_mes, values="sales", names="month",title="Distribuição de Vendas por Mês")
grafico_pizza.show()
```

Visualizar sazonalidade

```python
# Agrupar vendas por dia da semana
tabela_weekday_sales = tabela.groupby("day_of_week")["sales"].sum().reset_index()

# Ordenar os dias da semana corretamente
order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
tabela_weekday_sales["day_of_week"] = pd.Categorical(tabela_weekday_sales["day_of_week"], categories=order, ordered=True)
tabela_weekday_sales = tabela_weekday_sales.sort_values("day_of_week")

# Criar gráfico de barras
grafico_barras_semana = px.bar(tabela_weekday_sales, x="day_of_week", y="sales", title="Vendas por Dia da Semana", text_auto=True)
grafico_barras_semana.show()
```

Segmentar clientes - Analisar o perfil dos clientes, como idade e gênero

```python
# Criar um histograma da idade dos clientes
grafico_histograma_idade = px.histogram(tabela, x="customer_age", nbins=20, title="Distribuição de Idade dos Clientes")
grafico_histograma_idade.show()
```

```python
# Analisar vendas por gênero
tabela_genero_vendas = tabela.groupby("customer_gender")["sales"].sum().reset_index()

# Criar gráficos de barras
grafico_barras_genero = px.bar(tabela_genero_vendas, x="customer_gender", y="sales", title="Vendas por gênero", text_auto=True)
grafico_barras_genero.show()
```

# Passo 3: Aplicação de técnicas de Machine Learning

Devido a IA necessitar de números para aprender, realizo a transformação dos tipos de valores que vamos utilizar. É utilizado para esse fim o "LabelEncoder" da biblioteca scikit-learn.

```python
from sklearn.preprocessing import LabelEncoder

# customer_gender
codificador_genero = LabelEncoder()
tabela["customer_gender"] = codificador_genero.fit_transform(tabela["customer_gender"])

# category
codificador_categoria = LabelEncoder()
tabela["category"] = codificador_categoria.fit_transform(tabela["category"])

# unit_price
codificador_preco = LabelEncoder()
tabela["unit_price"] = codificador_preco.fit_transform(tabela["unit_price"])

# region
codificador_regiao = LabelEncoder()
tabela["region"] = codificador_regiao.fit_transform(tabela["region"])

# device
codificador_device = LabelEncoder()
tabela["device"] = codificador_device.fit_transform(tabela["device"])

# month
codificador_mes = LabelEncoder()
tabela['month'] = codificador_mes.fit_transform(tabela["month"])

# day_of_week
codificador_dia_semana = LabelEncoder()
tabela["day_of_week"] = codificador_dia_semana.fit_transform(tabela["day_of_week"])
```

```python
# Variável alvo - Serve como o parâmetro "Y" do aprendizado
alvo = tabela["sales"]

# Selecionar as variáveis que serão melhores utilizadas para o aprendizado de máquina - Serve como o parâmetro "X"
caracteristicas = tabela.drop(columns=["order_id", "date", "customer_id", "product_id", "product_name", "promo", "payment_method", "delivery_status", "return_status", "order_channel", "discount_code"])

# Importar modelo da inteligencia artificial para realizar o treino e teste
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(caracteristicas, alvo)
```

```python

# Escolha de modelo adequado - Selecionar 2 para realizar a verificação e ver qual se encaixa melhor
from sklearn.linear_model import LinearRegression # Regressão linear
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Criar o treino do 1º modelo
regressão_linear = LinearRegression()
regressão_linear.fit(x_train, y_train)

# Fazer previsões
previsao_regressao_linear = regressão_linear.predict(x_test)

# Avaliar o modelo calculando métricas de erro
import numpy as np
mae = mean_absolute_error(y_test, previsao_regressao_linear)
mse = mean_squared_error(y_test, previsao_regressao_linear)
rmse = np.sqrt(mse)

display(f"Erro médio absoluto: {mae:.2f}")
display(f"Erro quadrado médio: {mse:.2f}")
display(f"R.M.S.E: {rmse:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Criar o treino do 2º modelo
random_forest = RandomForestRegressor(n_estimators=100, random_state=42)

# Treinar o modelo
random_forest.fit(x_train, y_train)

# Fazer previsões
previsao_random_forest = random_forest.predict(x_test)

# Avaliação
mae_rf = mean_absolute_error(y_test, previsao_random_forest)
mse_rf = mean_squared_error(y_test, previsao_random_forest)
rmse_rf = np.sqrt(mean_squared_error(y_test, previsao_random_forest))

display(f"Erro médio absoluto (Random Forest): {mae_rf:.2f}")
display(f"Erro quadrado médio (Random Forest): {mse_rf:.2f}")
display(f"R.M.S.E (Random Forest): {rmse_rf:.2f}")
```

Visualizando as previsões

```python
import plotly.graph_objects as go

# Criar gráfico com valores reais e previstos
fig = go.Figure()
fig.add_trace(go.Scatter(y=y_test.values, mode='lines', name="Vendas Reais"))
fig.add_trace(go.Scatter(y=previsao_random_forest, mode='lines', name="Vendas Previstas (Random Forest)"))

fig.update_layout(title="Previsão de Vendas",
                  xaxis_title="Amostras",
                  yaxis_title="Vendas",
                  legend_title="Legenda")

fig.show()
```
